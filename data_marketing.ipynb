```python
#Transformer les dates

# Pour les clients
customers['first_purchase'] = pd.to_datetime(customers['first_purchase'])
customers['last_purchase'] = pd.to_datetime(customers['last_purchase'])

# Pour les transactions
transactions['invoice_date'] = pd.to_datetime(transactions['invoice_date'])

print("Datetime now exploitables !")
```

### Analyse & doublon du dataset

```python
# Analyse rapide du dataset
datasets = {"Clients": customers, "Transactions": transactions, "Campagnes": campaigns}

for nom, df in datasets.items():
    print(f"\n--- Analyse : {nom} ---")
    print(f"Lignes : {df.shape[0]} | Colonnes : {df.shape[1]}")

    # Vérification des doublons pour verifier lunicité
    nb_doublons = df.duplicated().sum()
    if nb_doublons > 0:
        print(f"{nb_doublons} lignes en double détectées !")
    else:
        print("propre propre propre !!!!")
```

Hypothèse : il se pourrait que le système ai pu valider 2 fois la même ligne si l'utilisateur rafraîchit sa page ou s'il y a un bug réseau ce qui expliquerait les doublons. Pour ne pas prendre de risques, nous allons les supprimer afin d'éviter de surestimer le CA de l'entreprise.

```python
# Nettoyer les doublons et corriger les ids

# On supprime les 34 522 doublons des transactions
transactions = transactions.drop_duplicates()

# On convertit l id client en entier
transactions['customer_id'] = transactions['customer_id'].astype('Int64')

print(f" Doublons supprimés. Nouveau volume : {len(transactions)} lignes.")
```

### Valeurs manquantes

```python
# Checkpoint des valeurs manquantes en %
def check_missing(df):
    missing_pct = (df.isnull().sum() / len(df)) * 100
    # On ne garde que les colonnes qui ont des manques pour y voir clair
    return missing_pct[missing_pct > 0].round(2)

print("Pourcentage de valeurs manquantes :")
print(f"Clients :\n{check_missing(customers)}\n")
print(f"Transactions :\n{check_missing(transactions)}\n")
print(f"Campagnes :\n{check_missing(campaigns)}")
```

Hypothèse : Nous remarquons que plus de 22% de customer_id n'est pas reliée à un client identifié, ceci est énorme, il se peut qu'on relève des bug dans le trackind de l'id client au moment de la vente.

*   Nous remarquons aussi que product_name & quantity sont  faible en proportions mais c'est étrange à creuser !



Nous avons observés une quantité de 50 000 clients enregistrés dans les données clients. Le volume de transactions s'élève à 1 837 137 transactions sur une période allant de 2007 à 2011. C'est un dataset qui couvre 5ans d'activités.

```python
# Le contrôle final

print("Rapport de qualité après nettoyage des données :\n")

print("Vérification des types:")
check_types = {
    "Transactions (Date)": transactions['invoice_date'].dtype,
    "Transactions (ID)": transactions['customer_id'].dtype,
    "Clients (Date)": customers['first_purchase'].dtype
}
for info, dtype in check_types.items():
    print(f" - {info}: {dtype} ")

# Vérification des doublons
print(f"\n Doublons restants dans Transactions : {transactions.duplicated().sum()}")

# État des Valeurs Manquantes (NaN)
print("\n Valeurs manquantes :")
def missing_report(df, name):
    pct = (df.isnull().sum() / len(df)) * 100
    print(f" --- {name} ---")
    print(pct[pct > 0] if pct.sum() > 0 else "Aucun vide !")

missing_report(transactions, "Transactions")
missing_report(customers, "Clients")

# Aperçu final
print("\n Aperçu des 3 premières lignes de Transactions :")
display(transactions.head(3))
```

**1 Analyse et nettoyage de transactions.csv**

Les customer_id manquants: que représentent ces lignes ? Faut-il les supprimer ou les conserver ?

**Analyse** : Environ 22,8% des transactions n'ont pas d'identifiant client. Cela correspond généralement à des ventes "Invités" (Guest Checkout) ou des erreurs techniques lors de l'export du CRM.

**Décision** : Nous conservons ces lignes pour ne pas sous-estimer le chiffre d'affaires global et la saisonnalité. Cependant, elles seront exclues lors de la segmentation comportementale (RFM) car l'identité client est manquante.



```python
# Identification des manquants
missing_cust = df_trans['customer_id'].isnull().sum()
print(f"Transactions sans customer_id : {missing_cust} ({missing_cust/len(df_trans):.1%})")
```

Les quantity négatives: d'où viennent-elles ? Quel impact sur les agrégations futures ?

**Analyse** : Les quantités négatives proviennent des retours clients, des annulations de commandes ou des ajustements d'inventaire.

**Décision** : Elles doivent être conservées pour le calcul du CA net (vision comptable), mais elles doivent être isolées pour ne pas biaiser le calcul du panier moyen marketing.

```python
# Identification des retours
print(f"Nombre de lignes avec quantité < 0 : {(df_trans['quantity'] < 0).sum()}")
```

Les unit_price à zéro: anomalie ou cas métier légitime ?


**Analyse** : Un prix nul peut être légitime (cadeaux promotionnels, échantillons, garanties) ou une anomalie (données manquantes, erreurs d'interface).

**Décision** : Nous les conservons s'ils sont associés à des noms de produits réels, mais nous les documentons comme étant des opérations marketing sans valeur monétaire directe.

```python
# Aperçu des produits à 0€
print("Exemples de produits à prix nul :")
print(df_trans[df_trans['unit_price'] == 0]['product_name'].unique()[:10])
```

Les product_code atypiques: identifiez les codes non-produits (frais de port, ajustements comptables, etc.)

Calculez line_total = quantity × unit_price. Vérifiez la cohérence.


**Analyse** : Certains codes comme POST (frais de port) ou D (Discount) ne sont pas des produits physiques mais des lignes de service ou d'ajustement.

**Décision** : Nous créons la colonne line_total pour vérifier la cohérence. Les lignes avec des quantités NaN sont supprimées car elles empêchent tout calcul fiable.

```python
# Suppression des lignes sans quantité pour la cohérence
df_trans = df_trans.dropna(subset=['quantity'])

# Calcul de la ligne totale
df_trans['line_total'] = df_trans['quantity'] * df_trans['unit_price']

# Identification des codes de service (contiennent des lettres et ne sont pas des variantes)
atypical_codes = df_trans[df_trans['product_code'].str.contains('^[a-zA-Z]+$', na=False)]['product_code'].unique()
print(f"Codes atypiques identifiés : {atypical_codes}")
```

## Etape III - Elyna

***Saisonnalité :*** **Imane**

**Évolution du CA mensuel:**

```python
import matplotlib.pyplot as plt

# Préparation : on s'assure que les dates et le CA sont bien calculés
transactions['invoice_date'] = pd.to_datetime(transactions['invoice_date'])
transactions['month_year'] = transactions['invoice_date'].dt.to_period('M').astype(str)
transactions['line_total'] = transactions['quantity'] * transactions['unit_price']

# Calcul du CA mensuel
ca_mensuel = transactions.groupby('month_year')['line_total'].sum()

# Graphique
plt.figure(figsize=(14, 5))
ca_mensuel.plot(kind='line', marker='o', color='#2ecc71', linewidth=2)
plt.title('Évolution du Chiffre d\'Affaires Mensuel', fontsize=14)
plt.ylabel('CA (€)')
plt.xlabel('Mois')
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

L'analyse du graphique révèle une tendance globale fortement croissante sur la période observée. L'activité n'est pas linéaire : on observe une accélération marquée du chiffre d'affaires à mesure que le temps progresse. Le volume d'affaires est soumis à des cycles mensuels significatifs, caractérisés par une croissance organique interrompue par des pics de consommation spécifiques. Cette dynamique indique un business en pleine expansion avec une saisonnalité qui s'accentue au fil des mois.

**Analyse des pics saisonniers et Implications Marketing:**

```python
# Identification des mois records pour confirmer les pics
top_3_mois = ca_mensuel.sort_values(ascending=False).head(3)

print("--- ANALYSE DES PICS DE VENTES ---")
for mois, valeur in top_3_mois.items():
    print(f"Mois : {mois} | Chiffre d'Affaires : {valeur:,.2f} €")

# Visualisation ciblée sur les pics
import matplotlib.ticker as ticker
fig, ax = plt.subplots(figsize=(12, 5))
ca_mensuel.plot(kind='bar', color=['#3498db' if x not in top_3_mois.values else '#e74c3c' for x in ca_mensuel.values], ax=ax)
plt.title('Identification des Pics de Saisonnalité (en rouge)', fontsize=14)
plt.ylabel('CA (€)')
plt.xticks(rotation=45)
plt.show()
```

L'examen des données confirme la présence de **pics saisonniers majeurs en novembre et décembre**, avec un volume de ventes nettement supérieur à la moyenne annuelle. Cette concentration de l'activité sur le dernier trimestre suggère une forte dépendance aux événements commerciaux de fin d'année (Black Friday et fêtes de Noël).

**Implications marketing stratégiques :**

**Planification et Logistique
:** Une phase d'anticipation
est cruciale dès le mois d'octobre pour sécuriser les stocks et optimiser la chaîne logistique afin d'absorber la hausse de la demande.

**Investissement Publicitaire :** Les budgets d'acquisition doivent être intensifiés juste avant ces pics (octobre/novembre) pour maximiser la visibilité au moment où l'intention d'achat est au plus haut.

**Cycle de Vie Client :** Puisque ces pics attirent de nombreux profils "saisonniers", une campagne de réactivation ou un programme de fidélité spécifique doit être lancé dès janvier pour convertir ces acheteurs ponctuels en clients récurrents et limiter le taux de churn post-fêtes.

```python
import seaborn as sns

# 1. On identifie les deux années qui ont le plus de transactions
top_years = transactions['year'].value_counts().nlargest(2).index.sort_values()
print(f"Analyse focalisée sur les années principales : {top_years.tolist()}")

# 2. On filtre le DataFrame pour ne garder que ces deux années
df_compare = transactions[transactions['year'].isin(top_years)]

# 3. Création du pivot pour ces deux années uniquement
pivot_ca = df_compare.pivot_table(index='month_num', columns='year', values='line_total', aggfunc='sum')

# 4. Visualisation
plt.figure(figsize=(12, 6))
sns.lineplot(data=pivot_ca, markers=True, dashes=False, palette='viridis', linewidth=2.5)

plt.title(f'Stabilité de la Saisonnalité : {top_years[0]} vs {top_years[1]}', fontsize=14)
plt.xticks(range(1, 13), ['Jan', 'Fév', 'Mar', 'Avr', 'Mai', 'Juin', 'Juil', 'Août', 'Sept', 'Oct', 'Nov', 'Déc'])
plt.ylabel('Chiffre d\'Affaires (€)')
plt.xlabel('Mois de l\'année')
plt.legend(title='Année')
plt.grid(True, alpha=0.3)
plt.show()
```

Oui, la saisonnalité est globalement **stable**, malgré une chute visible en novembre de la deuxième année.

**Pourquoi la chute en novembre ?**  Ce n'est pas une baisse de ventes réelle, mais un **biais de collecte** : les données de la deuxième année s'arrêtent au milieu du mois de novembre. Le mois est donc "incomplet", ce qui explique pourquoi le chiffre d'affaires paraît plus bas que l'année précédente.

**Stabilité :** Si on regarde les 10 premiers mois (Janvier à Octobre), les deux années suivent exactement le même mouvement.

**Conclusion :** Les habitudes d'achat se répètent bien d'une année sur l'autre. Il faut juste faire attention à cette fin de dataset incomplète pour ne pas tirer de fausses conclusions marketing.

***Relations entre variables :*** **Imane**

**Récence vs Montant dépensé**

```python

# Graphique de dispersion (Scatter Plot)
plt.figure(figsize=(10, 6))
sns.scatterplot(data=customers, x='recency_days', y='total_spent', alpha=0.5, color='orange')
plt.title('Relation entre Récence et Montant Total Dépensé')
plt.xlabel('Récence (Jours depuis le dernier achat)')
plt.ylabel('Montant Total (€)')
plt.show()

# Calcul de la corrélation de Pearson
print(f"Corrélation de Pearson : {customers['recency_days'].corr(customers['total_spent']):.2f}")
```

La corrélation **n'est pas linéaire** (-0.12 ).

**Observation :** On constate une concentration de points pour les faibles récences, mais le montant dépensé ne diminue pas de manière régulière avec le temps.

**Connexion cours EDA :** La relation est non-monotone. Les clients très récents ont un fort potentiel, mais certains clients très anciens possèdent également une valeur historique élevée. Une simple corrélation linéaire ne suffit pas à capturer la valeur d'un client.

**Fréquence vs Panier Moyen (Avg Basket)**

```python
# Graphique pour voir si la fréquence augmente le panier moyen
plt.figure(figsize=(10, 6))
sns.regplot(data=customers, x='n_orders', y='avg_basket',
            scatter_kws={'alpha':0.3}, line_kws={'color':'red'})

# On limite les axes pour mieux voir la masse de clients (évite les outliers)
plt.xlim(0, customers['n_orders'].quantile(0.98))
plt.ylim(0, customers['avg_basket'].quantile(0.98))

plt.title('Fréquence (Nombre de commandes) vs Panier Moyen')
plt.xlabel('Nombre de commandes')
plt.ylabel('Panier Moyen (€)')
plt.show()
```

Non, les clients fréquents ne dépensent pas forcément plus à chaque visite.

La ligne rouge (tendance) est presque plate.

Cela signifie que la fidélité (fréquence) ne fait pas augmenter la taille du panier moyen. Un client fidèle achète plus souvent, mais pour des montants similaires à chaque fois.

**Segments "naturellement visibles"**

```python
# Matrice de corrélation pour visualiser les liens globaux
plt.figure(figsize=(8, 6))
sns.heatmap(customers[['n_orders', 'total_spent', 'avg_basket', 'recency_days']].corr(),
            annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Matrice de corrélation des métriques CRM')
plt.show()
```

D'après les analyses précédentes, trois segments se distinguent clairement :

**Les "One-shot" (30,3%) :** Le groupe massif qui ne commande qu'une seule fois.

**Les "Champions" :** Clients avec une fréquence et un montant très élevés (le moteur de la croissance).

**Les "Anciens Gros Clients" :** Clients à fort montant cumulé mais avec une récence élevée (clients inactifs à réactiver en priorité).
